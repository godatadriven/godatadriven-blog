```python
import pandas as pd
import numpy as np
import numpy.random as rnd
from functools import partial
```

```python
NUM_PARTS = 20
NUM_ROWS = 600000000
NUM_SAMPLES = 10000

HIGH = 10000000
LOW = 20

CENTROIDS = [
    (10.0, 10.0),
    (100.0, 10.0),
    (100.0, 100.0),
    (10.0, 100.0),
]
```

```python
def generate_row(uid, high_card, low_card):
    c = CENTROIDS[int(rnd.rand() * 4)]
    return (
        uid,
        int(rnd.rand() * high_card),
        int(rnd.rand() * low_card),
        int(rnd.pareto(6.0) * high_card),
        int(rnd.randn() * high_card),
        rnd.randn() + c[0],
        rnd.randn() + c[1]
    )

def generate_rows(start, n):
    for x in xrange(start, start + n):
        yield generate_row(start * n + (x - start), HIGH, LOW)
```

```python
rdd = sc\
.parallelize(range(NUM_PARTS), numSlices=NUM_PARTS)\
.flatMap(partial(generate_rows, n=NUM_ROWS / NUM_PARTS))\
.map(lambda t: '|'.join(map(str, t)))\
.saveAsTextFile('/mnt/raid/numbers.txt')
```

```python
parsers = (int, int, int, int, int, float, float)
# Read back the file, because using the original RDD directly might partially re-generate random numbers
data = sc\
.textFile('/mnt/raid/numbers.txt/part*')\
.map(lambda line: map(unicode.strip, line.split('|')))\
.map(lambda d: [parser(d) for parser, d in zip(parsers, d)])

df = pd.DataFrame(
    data.sample(False, float(NUM_SAMPLES) / NUM_ROWS).collect(),
    columns=('uid', 'hc_unif', 'lc_unif', 'hc_pareto', 'hc_norm', 'x', 'y'))

df.head()
```
